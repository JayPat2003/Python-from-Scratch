## if statement


````python
# Python program to express a digit in a word

num = 1
if num == 1:
  print("One")
````

## if-else statement

The if..else statement executes a group of statements when a consition is True. Otherwise, it will execute another group of statements.

The syntax of if...else statement


```
if condition:
  statements1
else:
  statements2
```
If the 'If' condition is true then only the statement1 will be executed or else statement2 will be executed.

````python
# Python Program to know whether a number is even or odd
num=int(input("Enter a number: "))
if num % 2 == 0:
  print("Even")
else:
  print("Odd")
````

````python
# Python program to test whether the given number is between 1 and 20
num = int(input("Enter a number: "))
if num >= 1 and num <= 20: # this is compound condition
  print("Yes")
else:
  print("No")
````

## While loop

It is useful to execute a group of statements several times repeatedly depending upon whether the condution is True or False.

```
while condition:
  statements
```

The statements will be executed until and unless the condition will be True.

The While loop will be itertated until the condition becomes False.

````python
#Python program to display numbers from 1 to 10
x = 1
while x <= 10:
  print(x)
  x += 1
````

````python
x = 1
while x <= 5:
  print(x)
  x += 1
  print('End')
````

````python
# Python Program to display even numbers between 100 and 120
x = 100
while x >= 100 and x <= 120:
  if x % 2 == 0:
    print(x)
    x += 2
````

````python
# Write Python program for even numbers between m and n
m , n = map(int, input().split(','))
x = m
if x % 2 != 0:
  x += 1
while x >= m and x <= n:
  print(x)
  x += 2
````

## For Loop



*   It is used to iterate over the elements of a sequence.
* It is generally used when we know the till what range we want to iterate that condition.


```
for var in sequence:
  statements
```
* First, the sequence is assigned to the variable written after the 'for' and then the statements are executed.
* Next, the 2nd element sequence  is assigned to the variable and then the statements are executed 2nd time.




````python
# A python program to display characters of a string using for loop
str = "Python"
for c in str:
  print(c)
````

````python
#display each character from a string
str = 'Python'
n = len(str) # gives the total number of elements in the sequence like a string, list or tuple
for i in range(n):
  print(str[i])
````



*   The range() object also known as range() function is useful to provide a sequence of numbers.

* We can also include stepsize for the increment in value of variable at each step. For example : range(1,10,2), it will give output as 1,3,5,7,9.

````python
# Python program to display and find the sum of a list of numbers using while loop
list = [10,20,30,40]
sum = 0
i = 0
while i < len(list):
  print(list[i])
  sum += list[i]
  i += 1
print("Sum =", sum)
````

## Nested Loops



We can write one loop inside an another loop and such loops are called as **nested loops**.

````python
for i in range(3):
    for j in range(4):
        print('i =', i, '\t', 'j =', j)
````

When the outer loop is executed once, the inner loop is executed 4 times. It means that when i value is 0, j values will change from 0 to 3 and hence the `print()` funnction will display the above output.

````python
# Python Program to display stars in a right angled triangular form using nested loops
# Row number = number of stars in that row

for i in range(1,11): # display 10 rows
    for j in range(1, i+1): # no. of stars = row number
        print('* ', end='')
    print() # throws the cursor into the next line
````

````python
# Python Program to display stars in a right angled triangular form using single loop
for i in range(1,11):
    print('* '*(i)) # repeat star for i times
````

````python
# Python Program to display stars in equilateral traingle using single loop
n = 40
for i in range(1,11):
    print(' '*n, end='') # repeat space for n times
    print('* '*(i)) # repeat star for i times
    n -= 1
````

````python
# Python program to display numbers from 1 to 100 in a proper format
for x in range(1,11):
    for y in range(1,11):
        # specify the column sixr by using replacement field {} and format() method
        print('{:5}'.format(x*y), end='') # each column size is 5
    print()
````

**Why else suite is useful?**

Sometimes, we write programs where searching for an element is done in the sequence. When the element is not found - we can indicate that in the else suite easily.

````python
# Python program to search for an element in the list of elements
grp1 = [2,9,12,90]
search = int(input('Enter the element to search: '))
for element in grp1:
    if search == element:
        print('Element found in group 1')
        break # to come out of the loop
else:
    print('Element not found in the group 1') # this is else suite
````

## The break Statement

* Break statement can be used inside a for loop or while loop to come out of the loop.
* When the `break` is executed the Python interpreter jumps out of the loop to process the next statement in the program.

````python
x = 10
while x >= 1:
    print('x=', x)
    x -= 1
print('Out of loop')
````

Python Program to display numbers from 10 to 6 and break the loop when 5 appeears

````python
x = 10
while x >= 1:
    print('x is ', x)
    x -= 1
    if x == 5:
        break
print('out of loop')
````

## Continue Statement

* It is used in a loop to go back to the beginning of the loop.
* When continue is executed, the next repetition will start. When continue is executed, the subsequent statements in the loop are not executed.

````python
x = 0
while x < 10:
    x += 1
    if x > 5:
        continue
    print('x is ',x)
print('Out of Loop')
````

In the following above code, the while loop repeats for 10 times from 0 to 9 and everytime the x value is displayed too but when x is greater than 5, the 'continue' is executed that makes the Python Interpreter to go back to the beginning of the loop.

## Pass Statement

We use the pass when we need a statement syntatically but we do not want to do any operation. It can be used to inform the Python Interpreter to not to do anything when we are not interested in the result.

````python
x = 0
while x < 10:
    x += 1
    if x == 6:
        pass
    print('x is ', x)
print('Out of loop')
````

Python program to display a series of Prime Numbers

````python
max = int(input('Enter the number till where you want the prime numbers : '))

for num in range(2,max+1):
    for i in range(2, num):
        if num % i == 0:
            break #then it is not prime, hence go back and check next
    else:
        print(num)
````

````python
# program to check whether the following number is prime or not
n = 17

# A prime number is greater than 1 and has no divisors other than 1 and itself.
if n > 1:
    for i in range(2, int(n ** 0.5) + 1):  # Check divisors only up to sqrt(n)
        if n % i == 0:
            print('The number is not prime')
            break
    else:
        print('The number is prime')
else:
    print('The number is not prime')
````

````python
# Program to generate Fibonacci Series

n = int(input('How many Fibonaccis? '))
f1 = 0
f2 = 1
c = 2

if n == 1:
    print(f1)
elif n == 2:
    print(f2)
else:
    print(f1, '\n', f2 , sep = '')
    while c < n:
        f = f1 + f2
        print(f)
        f1 = f2
        f2 = f
        c += 1
````
